## Project submission: Navigation Project

# Learning algorithm

This environment was solved using a Deep Q Network. The agent was identical to the one used for an earlier exercise. 

The agent uses a simple three layer fully connected neural network to determine the Q-value. The layers are connected using ReLu activation functions. There are two separate networks with identical structures. 

The parameters used for the agent are:  
Buffer size: 10000  
Batch size: 64  
Discount factor (Gamma): 0.99 
Learning rate: 5e-4  
Network is updated every 4 batches  

# Plot of Rewards

Environment solved to a score of 16.01 in 845 episodes

# Ideas for Future Work

The performance of the current model is reasonably good. It exceeds the benchmark of 13 with less than 1000 episodes. However, there are many exciting directions in which this project can go:
1. (Course suggested) Train the model on actual image (pixels)
2. (Alternate algorithms) Double DQN networks
3. (Alternate algorithms) Dueling DQNs
4. (Existing algorithm) Tune hyper parameters of network used in the agent. Currently the model being used is a simple three layer fully connected network. It'll be interesting to explore what happens as the number of layers, type of layers and type of activation function is changed.
